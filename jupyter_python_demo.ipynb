## If statements

If statements allow you to take actions only if a condition is met they are of the form

```
if expression:
    statements
else:
    more statements
```

(note the indentation)

`expression` is the test to see if the condition is met

conditions can be either `True` or `False`

The `==` operator checks if two things are equal

````python
1==1
````

````python
1==2
````

````python
# Add only odd numbers

numbers=range(10)
sum=0

for x in numbers:
    if x%2==1:
        print("Adding",x,"to",sum)
        sum=x+sum
        print("sum =",sum,"\n")
    else:
        print("Skipping",x)
````

* `value%2` returns 0 (`False`) if `value` is divisible by 2
* `value%2` returns 1 (`True`) if `value` is not divisible by 2
* `\n` is a "newline" character. Helpful for readability

You can combine conditions with `and` and `or`

````python
# Both conditions must be True
1==1 and 1==2
````

````python
#Only one needs to be True
1==1 or 1==2
````

````python
# Add only odds that are divisible by 3

numbers=range(10)
sum=0

for x in numbers:
    if x%2==1 and x%3==0:
        print("Adding",x,"to",sum)
        sum=x+sum
        print("sum =",sum,"\n")
    else:
        print("Skipping",x)
````

## While loops

While loops will execute as long as a particular condition is `True`

````python
# Add every number up to a limit

x=0
max_value=15
sum=0

while x<max_value:
    print("Adding",x,"to",sum)
    sum=x+sum
    print("sum =",sum,"\n")
    
    # Don't forget to update x or loop will never stop!
    x=x+1
````

## Creating your own functions

* A set of actions that you want to do frequently
* They can take arguments
    * Arguments can have default values
* They can modify arguments

Let's say we want to find every Friday the 13th in a given year:
* Create function to check if a particular day is Friday the 13th
* Loop over entire year and check each day

````python
import datetime as dt
````

````python
def isFridaythe13(day):
    
    
    # If it's friday, day.weekday=4
    # Monday==0, ..., Sunday==6
    weekday=day.weekday()
    
    # If weekday.day==13, it's the 13th of the month
    day_number = day.day
    
    #Is it a Friday AND the 13th?
    if weekday==4 and day_number==13:
        return True
    else:
        return False
````

Functions that return `True` or `False` traditionally have names that begin with `is` (like it's asking a question)

````python
# Find every Friday the 13th in a given year

year=2022

months=["January","February","March","April","May","June",
        "July","August","September","October","November","December"]

first_day=dt.date(year,1,1)
one_day=dt.timedelta(days=1)

current_day=first_day
while current_day.year==year:
    
    if isFridaythe13(current_day):
        # Subract current_dat.month by one to get proper months[] index
        print("Friday the 13th on",months[current_day.month-1], year)
        
    # Don't forget to update the day!
    # This line always runs since it's not indented with the 'if' statement
    current_day=current_day+one_day
````

There are a couple reasons to use a function:
   * Breaks code up into bite-sized chunks
   * Makes code more readable
   * Easier to do a complicated action over and over

### Pendulums
![pendulum](images/pendulum_y.svg)

Pendulum with mass $m$, on rope of length $L$ swings back and forth because of gravity

A physicist might create a diagram like this
![pendulum](images/pendulum_y_forces.svg)

and then apply some geometry
![pendulum](images/pendulum_y_forces_all.svg)

Conclusion:
* Tension hold the rope up (mass is always distance $L$ away from top)
* Unbalanced force, $mg sin\theta$ keeps the pendulum moving

Equation of motion that describes this situation is:

$$  \frac{d^2 y}{dt^2} = -\frac{g}{L}sin y $$

Angular acceleration depends on angular position.

It turns out that this problem is unsolvable (no function, $y(t)$, can satisfy these constraints)

Examples of functions

$$ y(t) = t^2+4t+3$$
$$ y(t) = 5e^t$$

Some people have noticed that when $y$ is small, $sin y \approx y$ so the equation of motion can be written

$$  \frac{d^2 y}{dt^2}= -\frac{g}{L} y $$

(This is solveable)

We want to know:
1. Is this approximation justified?
1. When does the approximation breakdown?
1. If we use this approximation, what behavior are we missing out on?

We can answer all these questions in Python!

Let's say we want 10% accuracy as defined by

$$ \%\Delta= \frac{x -y}{y} \times 100\% $$

````python
diff=(x-y)/y
````

There's a problem because some of the values in `y` are zero.

We can ignore those cases to avoid dividing by zero

````python
# Divde x-y but ignore when y=0
# to avoid divide by zero errors

diff=np.true_divide(x-y,y,where=y!=0)

# Returns list of array indicies where dif > 0.1
over_10_percent=np.argwhere(diff > 0.01)

# Find the first one, subtract by 1
last_under_10=over_10_percent[0,0]-1
print("When x~%.2f rad (%.2f deg) diff is %.2f%%" %(x[last_under_10],x[last_under_10]*180/pi,100*diff[last_under_10]))
````

````python
import scipy.integrate as integrate
````

````python

# Set length of pendulm and strength of gravity
gL=9.81/10

# Define angular velocity and angular acceleration
def deriv_ivp_approx(t,y):
    return    [y[1],-gL*y[0]]
    
def deriv_ivp(t,y):
    return    [y[1],-gL*sin(y[0])]
````

````python
min_t=0
max_t=70
init_vals=[90*pi/180,0] #inital angular postition and velocity
t=np.linspace(min_t,max_t,2000) # time period of interest

solution = integrate.solve_ivp(deriv_ivp,[min_t,max_t],init_vals,t_eval=t)
solution_approx=integrate.solve_ivp(deriv_ivp_approx,[min_t,max_t],init_vals,t_eval=t)
````

````python
fig, axs=plt.subplots(2,2)
fig.dpi=150

# Top left
axs[0,0].plot(solution.t,solution.y[0]*180/pi)
axs[0,0].set_xlabel('Time (s)')
axs[0,0].set_ylabel('Angular position')
axs[0,0].set_title('"True" Solution')

# Top right
axs[0,1].plot(solution_approx.t,solution_approx.y[0]*180/pi,'tab:orange')
axs[0,1].set_title('Approximate Solution')


# Bottom left
axs[1,0].plot(solution.y[0],solution.y[1])
axs[1,0].set_aspect(1)
axs[1,0].set_xlabel('Angular position')
axs[1,0].set_ylabel('Angular velocity')


#Bottom right
axs[1,1].plot(solution_approx.y[0],solution_approx.y[1],'tab:orange')
axs[1,1].set_aspect(1)
axs[1,1].set_xlabel('Angular position')


legend = ax.legend()

plt.show()
````
